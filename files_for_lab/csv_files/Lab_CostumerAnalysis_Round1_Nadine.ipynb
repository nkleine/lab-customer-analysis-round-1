```python
#9 Perform the data cleaning operations mentioned so far in class:
```

```python
 #9a Delete the column education and the number of open complaints from the dataframe.
```

```python
filescomb= filescomb.drop(['education','number of open complaints'], axis=1)
filescomb.head()
```

```python
#9b Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
```

```python
filescomb['customer lifetime value']=list(map(lambda x: str(x).replace('%', ''),filescomb['customer lifetime value']))
filescomb.head(20)
```

```python
filescomb['customer lifetime value'] = [float(x) for x in filescomb['customer lifetime value'].values]
filescomb.head(20)
```

```python
filescomb.dtypes
```

```python
filescomb['customer lifetime value']=filescomb['customer lifetime value']/100
filescomb.head()
```
